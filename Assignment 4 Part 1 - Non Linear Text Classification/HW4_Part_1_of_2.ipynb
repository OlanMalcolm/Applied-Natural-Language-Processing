# Q1: Classification with LSTM [15pts]

A Long Short-Term Memory Network (LSTM) is a variant of the Recurrent Neural Network (RNN) that is able to deal with the vanishing gradient problem that we sometimes encounter in traditional RNNs. The vanishing gradient problem occurs when gradients become smaller and smaller as they backpropagate, leading to difficulties in learning long-term dependencies. LSTM's uses a cell state to remember information over arbitrary time intervals, and a combination of 3 gates, an input gate, an output gate and a forget gate, to regulate the flow of information into and out of the cell. By selectively keeping track of the relevant information in the cell state, the LSTM is able to learn useful long-term dependencies. The vanishing gradient problem is also alleviated because gradients are allowed to flow unchanged through the cell state.

We will be using a Bi-LSTM (Bidirectional LSTM) for classification. The architecture of our model looks like :

<p align="center"><img src="https://www.tensorflow.org/static/text/tutorials/images/bidirectional.png" width="75%" align="center"></p>

**NOTE**: It is important to note that we use the hidden state corresponding to the last word in the forward direction, which isn't necessarily the hidden state of the last time-step. For example if we have 3 sentences of lengths 10, 15 and 20, all 3 sentences will be padded up to 20 tokens during the pre-processing and vectorizing step (which has been implemented in the vectorize_batch( ) function below). We will be using the hidden state corresponding to the last word in each sentence (which will be
$h_{10}^f$, $h_{15}^f$ and $h_{20}^f$ respectively), not the hidden state of the last time-step (which is $h_{20}^f$ for all 3 sentences).

For more details on LSTM, please refer to class lectures.

### 1.1.1 : Pre-Processing Data [No Points]

```python
###############################
### DO NOT CHANGE THIS CELL ###
###############################

from torchtext.data import get_tokenizer
from torchtext.vocab import build_vocab_from_iterator
import torchtext

tokenizer = get_tokenizer("basic_english")

def build_vocabulary(datasets):
    for dataset in datasets:
        for text in dataset:
            yield tokenizer(text)

vocab = build_vocab_from_iterator(build_vocabulary([x_train]), min_freq=1, specials=["<UNK>"])
vocab.set_default_index(vocab["<UNK>"])

vocab_wos = build_vocab_from_iterator(build_vocabulary([x_train_wos]), min_freq=1, specials=["<UNK>"])
vocab_wos.set_default_index(vocab["<UNK>"])
```

### 1.2.1 : Create the Dataloaders [No Points]

```python
###############################
### DO NOT CHANGE THIS CELL ###
###############################

from torch.utils.data import DataLoader

max_words = max(map(len, x_train))

def vectorize_batch(batch):
    Y, X = list(zip(*batch))
    X = [vocab(tokenizer(text)) for text in X] ## Tokenize and map tokens to indexes
    X_len = [len(text) for text in X]
    X = [tokens+([0]* (max_words-len(tokens))) if len(tokens)<max_words else tokens[:max_words] for tokens in X] ## Bringing all samples to max_words length.
    return torch.tensor(X, dtype=torch.int32), torch.tensor(X_len), torch.tensor(Y)
```

### 1.2.3 : Train and Evaluate on the Clickbait dataset [5 Pts]

```python
###############################
### DO NOT CHANGE THIS CELL ###
###############################

# Use cuda if present
device = torch.device("cuda" if torch.cuda.is_available() else "cpu")
print("Device available for running: ", device)
```

```python
###############################
### DO NOT CHANGE THIS CELL ###
###############################

from lstm import LSTM
from tqdm import tqdm

NUM_CLASSES = 2

model = LSTM(vocab, num_classes=NUM_CLASSES)
model.to(device)
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.01)
N_EPOCHS = 5 if not SHORT else 1

model.train()
for epoch in range(N_EPOCHS):
    total_loss = 0.0
    for X, X_len, Y in tqdm(train_loader):
        X = X.to(device)
        Y = Y.to(device)
        outputs = model(X, X_len)
        loss = criterion(outputs, Y)

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        total_loss += loss.item()

    print("loss on epoch %i: %f" % (epoch, total_loss))
```

# Q2: Classification with LSTM + Attention [15pts]

A vanilla LSTM approach has a few potential issues:

1. It needs to compress all of the necessary information in a source sentence into a fixed-length vector. This may make it difficult for it to cope with long sentences, especially those that are longer than the sentences in the training corpus.
2. Lack of selective focus: LSTM's process sequential data step-by-step, but not all parts of the input sequence are equally relevant for the task at hand.

You will now explore an attention mechanism that offers a solution to both of these issues. Attention allows the model to focus on the most relevant parts of the input sequence at each step, giving more weight to important information. By selectively attending to different parts of the input, the model can capture long-range dependencies and contextual information more effectively.

There are several types of attention mechanisms that can be applied in NLP problems, but for this homework you will be implementing the Dot-Product Attention technique. It was proposed by [Luong et al.](https://arxiv.org/abs/1508.04025v5), who proposed a Global Attention Model, which would consider the hidden state at each time step. This model computes a variable-length alignment vector $\alpha$, which learns the relevant contextual information in the input sequence for the task at hand (in our scenario: sentence classification). Given the alignment vector as weights, the context vector is computed as the weighted average of the LSTM hidden states across all timesteps.

In this task, we will be implementing LSTM with Attention. The model architecture is shown below:

<p align="center"><img src="./data/images/q2_attention.png" width="75%" align="center"></p>

**NOTE:** The paper linked above proposes the Dot-Product Attention technique for a Machine-Translation task, which requires an encoder-decoder architecture (Many-Many). Since we are dealing with a classification problem (which requires a Many-One architecture), the raw alignment scores are computed a little differently.
1. The raw alignment scores are computed as $e_{i} =  a(h_{i}, h_{FINAL})$ $\forall i \in [1, 2, ... T]$ where $a$ is the alignment model (a dot product operation) and $h_{FINAL} = (h_n^f, h_1^b,)$ is the final hidden state of the LSTM obtained by concatenating the forward hidden state for the last word ($h_n^f$) with the backward hidden state for the first word ($h_1^b$). Also, $h_{i}$ represents the hidden state vector corresponding to the i'th time step, and it's made up of the forward and backward hidden state vectors at the i'th time step $h_{i} = (h_{i}^f, h_{i}^b)$.
2. The normalized attention weights are computed as $\alpha_{i} = \frac{\exp e_{i}}{\sum_{i=1}^T {\exp e_{i}}}$.
3. The context vector is computed using this formula: $c = \sum_{i=1}^T \alpha_i * h_{i}$
4. The context vector $c$ should be passed alongside the final hidden state of the LSTM $h_{FINAL} = (h_n^f, h_1^b,)$ to the Context Layer. The context layer is just another linear layer that needs to be initialized. You will need to determine its input and output sizes from the model architecture.


Please refer to lecture for more details.

You will be extending the LSTM model and incorporating attention on top of it.

## 4.4 :  Pre-process Dataset [No Points]

Run the below cells to vectorize the dataset for tokenizers and to intialize the dataloaders.

```python
###############################
### DO NOT CHANGE THIS CELL ###
###############################

### The code is used to vectorize the data for each batch as defined in the data loader.
### For data in the batch, the input tokens are encoded and
### then special tokens [CLS] (101) is added at the beginning and [SEP] (102) is added at the end.
### These tokens are added because the BERT model was pretrained with these tokens. So to get the same results for inference we need to add them.
### The input is padded with 0 if it is lesser than Max length.

MAX_LEN = 128

def vectorize_batch(batch):
    batch_input_ids = []
    batch_mask = []
    batch_token_type_ids = []
    batch_pos = []
    batch_ner = []

    for data in batch:
        target_pos = []
        target_ner = []
        inputs = []
        tokens = data['tokens']
        pos_tags = data['pos_tags']
        ner_tags = data['ner_tags']
        for i in range(len(tokens)):
            input = tokenizer.encode(tokens[i], add_special_tokens=False)
            input_len = len(input)
            target_pos.extend([pos_tags[i]] * input_len)
            target_ner.extend([ner_tags[i]] * input_len)
            inputs.extend(input)
        inputs = inputs[:MAX_LEN - 2]
        target_pos = target_pos[:MAX_LEN - 2]
        target_ner = target_ner[:MAX_LEN - 2]

        inputs = [101] + inputs + [102]
        target_pos = [0] + target_pos + [0]
        target_ner = [0] + target_ner + [0]

        mask = [1] * len(inputs)
        token_type_ids = [0] * len(inputs)

        padding_len = MAX_LEN - len(inputs)
        inputs = inputs + ([0] * padding_len)
        mask = mask + ([0] * padding_len)

        token_type_ids = token_type_ids + ([0] * padding_len)
        target_pos = target_pos + ([0] * padding_len)
        target_ner = target_ner + ([0] * padding_len)

        batch_input_ids.append(inputs)
        batch_mask.append(mask)
        batch_token_type_ids.append(token_type_ids)
        batch_pos.append(target_pos)
        batch_ner.append(target_ner)

    return torch.tensor(batch_input_ids, dtype=torch.long), torch.tensor(batch_mask, dtype=torch.long), torch.tensor(batch_token_type_ids, dtype=torch.long), torch.tensor(batch_pos, dtype=torch.long), torch.tensor(batch_ner, dtype=torch.long)
```
